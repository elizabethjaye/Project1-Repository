```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

csv = "../Resources/demographics.csv"
census_df = pd.read_csv(csv)
census_df.head()
```

```python
list(census_df)
```

```python
# X-Axis list
x_axis = census_df["Year"].tolist()
x_axis
```

```python
# EDUCATION PLOT

# Extract school levels as list
highschool = census_df["Number of Higher Ed Enrolled"].tolist()
somecollege = census_df["Education of Some College - No Degree Attained"].tolist()
associates = census_df["Education of Associate's Degree Attained"].tolist()
bachelors = census_df["Education of Bachelor's Degree Attained"].tolist()
graduate = census_df["Education of Graduate or Professional Degree Attained"].tolist()

# Define where bars stack on each other on y-axis
assoc_bot = [i+j for i,j in zip(highschool, somecollege)]
bach_bot = [i+j for i,j in zip(assoc_bot, associates)]
grad_bot = [i+j for i,j in zip(bach_bot, bachelors)]

# Single place to adjust bar width
width = .3

# Plots
edu_hs = plt.bar(x_axis, highschool, width, label="High School Diploma",color='yellow')
edu_some = plt.bar(x_axis, somecollege, width, bottom=highschool, label="Some College, no degree", color='orange')
edu_assoc = plt.bar(x_axis, associates, width, bottom=assoc_bot, label="Associates Degree", color='r')
edu_bach = plt.bar(x_axis, bachelors, width, bottom=bach_bot, label="Bachelors Degree", color='purple')
edu_grad = plt.bar(x_axis, graduate, width, bottom=grad_bot, label="Graduate Degree or higher", color='blue')

# Labeling
plt.xlabel("Years from US Census")
plt.ylabel("Number of People Enrolled in School")
plt.title('Highest Level of Education Attained')
plt.xticks(x_axis, ('2014', '2015', '2016', '2017'))
plt.legend(bbox_to_anchor=(1.6, 1))

plt.savefig("../PNG Plots/education.png")
plt.show()
# To Do list:
#  - Move Legend. Tightening bars? Move Legend outside of plot?
#  - Add percents and totals to bars
```

```python
# AGE PLOT

# Extract age ranges as list
agegrp1 = census_df["Age of 20 to 24 years"].tolist()
agegrp2 = census_df["Age of 25 to 34 years"].tolist()
agegrp3 = census_df["Age of 35 to 44 years"].tolist()
agegrp4 = census_df["Age of 45 to 54 years"].tolist()
agegrp5 = census_df["Age of 55 to 59 years"].tolist()
agegrp6 = census_df["Age of 60 to 64 years"].tolist()
agegrp7 = census_df["Age of 65 to 74 years"].tolist()
agegrp8 = census_df["Age of 75 to 84 years"].tolist()
agegrp9 = census_df["Age of 85 years and over"].tolist()

# Define where bars stack on each other on y-axis

agegrp3_bot = [i+j for i,j in zip(agegrp1, agegrp2)]
agegrp4_bot = [i+j for i,j in zip(agegrp3_bot, agegrp3)]
agegrp5_bot = [i+j for i,j in zip(agegrp4_bot, agegrp4)]
agegrp6_bot = [i+j for i,j in zip(agegrp5_bot, agegrp5)]
agegrp7_bot = [i+j for i,j in zip(agegrp6_bot, agegrp6)]
agegrp8_bot = [i+j for i,j in zip(agegrp7_bot, agegrp7)]
agegrp9_bot = [i+j for i,j in zip(agegrp8_bot, agegrp8)]

# Single place to adjust bar width
width = .3

# Plots
agegrp1_plt = plt.bar(x_axis, agegrp1, width, label="Age of 20 to 24 years", color='yellow')
agegrp2_plt = plt.bar(x_axis, agegrp2, width, bottom=agegrp1, label="Age of 25 to 34 years", color='orange')
agegrp3_plt = plt.bar(x_axis, agegrp3, width, bottom=agegrp3_bot, label="Age of 35 to 44 years", color='red')
agegrp4_plt = plt.bar(x_axis, agegrp4, width, bottom=agegrp4_bot, label="Age of 45 to 54 years", color="blue")
agegrp5_plt = plt.bar(x_axis, agegrp5, width, bottom=agegrp5_bot, label="Age of 55 to 59 years", color="lightgreen")
agegrp6_plt = plt.bar(x_axis, agegrp6, width, bottom=agegrp6_bot, label="Age of 60 to 64 years", color="pink")
agegrp7_plt = plt.bar(x_axis, agegrp7, width, bottom=agegrp7_bot, label="Age of 65 to 74 years", color="skyblue")
agegrp8_plt = plt.bar(x_axis, agegrp8, width, bottom=agegrp8_bot, label="Age of 75 to 84 years", color="magenta")
agegrp9_plt = plt.bar(x_axis, agegrp9, width, bottom=agegrp9_bot, label="Age of 85 years and over", color="brown")

# Labeling
plt.xlabel("Years from US Census")
plt.ylabel("Number of People")
plt.title('Number of People over the Age of 20 \n In Age Groups')
plt.xticks(x_axis, ('2014', '2015', '2016', '2017'))
plt.legend(bbox_to_anchor=(1.6, 1))

plt.savefig("../PNG Plots/age_groups.png")
plt.show()
# To Do list:
#  - Make bigger/fewer groups?
#  - Move Legend. Tightening bars? Move Legend outside of plot?
#  - Add percents and totals to bars
```
